```python
def compute_central_tendency(data, label):
    """
    Compute mean, median, and mode from scratch using numpy.
    No pandas .mean() or .median() shortcuts.
    
    data  : numpy array of float values (NaNs already removed)
    label : string name for printing
    """

#-----MEAN-----
# Formula: sum of all values divided by count
    mean_val = np.sum(data) / len(data)

 # Verify against numpy's built-in (they must match exactly)
    assert abs(mean_val - np.mean(data)) < 0.001, "Mean mismatch"


# -----MEDIAN-----
    # Formula: sort values, pick the middle one
    sorted_data = np.sort(data)
    n = len(sorted_data)

    if n % 2 == 1:
        # Odd count: exact middle index
        median_val = sorted_data[n // 2]
    else:
        # Even count: average the two middle values
        mid = n // 2
        median_val = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    
    assert abs(median_val - np.median(data)) < 0.001, "Median mismatch"


    # -----MODE----- 
    # Formula: the value that appears most often
    # np.unique returns every unique value AND how many times it appeared

    unique_vals, counts = np.unique(data, return_counts=True)
    mode_idx = np.argmax(counts)        # index of the highest count
    mode_val = unique_vals[mode_idx]
    mode_count = counts[mode_idx]

    # -----GAP-----
    # The mean-median gap tells you how skewed the distribution is
    gap = mean_val - median_val


    print(f"\n{'='*45}")
    print(f"  {label}")
    print(f"{'='*45}")
    print(f"  Mean   : {mean_val:.2f}")
    print(f"  Median : {median_val:.2f}")
    print(f"  Mode   : {mode_val:.2f}  (appeared {mode_count} times)")
    print(f"  Gap (mean - median): {gap:.2f}")


    if gap > 20:
        print(f"  >> Large gap — distribution is RIGHT-SKEWED")
        print(f"  >> A few extreme days are pulling the mean up")
    elif gap > 5:
        print(f"  >> Mild skew present")
    else:
        print(f"  >> Distribution is roughly symmetric")
```
