```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
df = pd.read_csv("../data.csv", encoding="cp1252")
df.head()
```

```python
delhi = df[df["state"] == "Delhi"]
print(f"Delhi rows: {len(delhi)}")
```

```python
state_pm25 = df.groupby("state")["pm2_5"].mean()
state_pm25 = state_pm25.dropna().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
state_pm25.plot(kind="bar")
plt.title("Average PM2.5 by State")
plt.xlabel("State")
plt.ylabel("Mean PM2.5 (µg/m³)")
plt.xticks(rotation=45, ha="right")
plt.savefig("pm25_by_state.png", dpi=150)
plt.tight_layout()
plt.show()
```

### PM2.5 by state

Delhi stands out clearly as the worst state, with an average PM2.5 of 95.11 µg/m³ - nearly 1.5x higher than second-place
Madhya Pradesh (65.06). Every other state is significantly lower, making Delhi a clear outlier at the national level.

```python
msno.matrix(df.sample(500, random_state=42))
plt.title("Missing Value Patterns — India AQI")
plt.tight_layout()
plt.savefig("null_matrix.png", dpi=150)
plt.show()
```

### Null matrix

Dark = missing. White = data present.

- pm2_5 is almost fully dark — 97.86% missing nationally.
- spm is mostly dark — over 54% missing.
- so2 and no2 have dark blocks in chunks — entire cities had no sensors, not random gaps.
- state and location are fully white — always filled in.

```python
pollutant_cols = ["so2", "no2", "rspm", "spm", "pm2_5"]
corr_matrix = df[pollutant_cols].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0)
plt.title("Pollutant Correlation Matrix — India AQI")
plt.tight_layout()
plt.savefig("pollutant_correlation.png", dpi=150)
plt.show()
```
